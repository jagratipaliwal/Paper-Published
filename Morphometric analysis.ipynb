```python
import numpy as np
import glob
import matplotlib.pyplot as plt
import skimage.io
import skimage.color
import skimage.filters


# load the image
image = skimage.io.imread("https://webapp.agron.ksu.edu/agr_social/lib/Filemanager/userfiles/06122020/Corn-leaf-diseases-in-Kansas-F02.jpg")

fig, ax = plt.subplots()
plt.imshow(image)
plt.show()
plt.imsave("test.jpg", image, )
```

```python
# convert the image to grayscale
gray_image = skimage.color.rgb2gray(image)

# blur the image to denoise
blurred_image = skimage.filters.gaussian(gray_image, sigma=0.5)

fig, ax = plt.subplots()
plt.imshow(blurred_image, cmap="gray")
plt.show()
plt.imsave("gray image.jpg", blurred_image, cmap="gray")
```

```python
# create a histogram of the blurred grayscale image
histogram, bin_edges = np.histogram(blurred_image, bins=256, range=(0.0, 1.0))

fig, ax = plt.subplots()
plt.plot(bin_edges[0:-1], histogram)
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixels")
plt.xlim(0, 1.0)
plt.show()
```

```python
# create a mask based on the threshold
t = 0.70



binary_mask = blurred_image > t

fig, ax = plt.subplots()
plt.imshow(binary_mask, cmap="gray")

plt.show()
plt.imsave("logo.jpg", binary_mask, cmap="gray")




```
